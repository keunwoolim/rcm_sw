```python
import numpy as np
import pandas as pd
import os
```

```python
distance_mean = np.zeros((4, 3))
iter = 0

for sample_style in ["spherical", "degenerate", "discrete"]:
    distance = np.zeros((4, 100))
    distance[0, :] = np.load(os.getcwd() + "\\distance_data\\approximate_distance_" + sample_style +"_2_500.npy")
    distance[1, :] = np.load(os.getcwd() + "\\distance_data\\approximate_distance_" + sample_style +"_2_1000.npy")
    distance[2, :] = np.load(os.getcwd() + "\\distance_data\\approximate_distance_" + sample_style +"_2_1500.npy")
    distance[3, :] = np.load(os.getcwd() + "\\distance_data\\approximate_distance_" + sample_style +"_2_2000.npy")
    distance_mean[:, iter] = np.mean(distance, axis = 1).T
    iter += 1

index = ["500", "1000", "1500", "2000"]
columns = ["spherical", "degenerate", "discrete"]
pd.DataFrame(distance_mean, index = index, columns = columns)
```

```python
distance_mean = np.zeros((4, 3))

for dimension in range (2, 6):
    distance = np.load(os.getcwd() + "\\distance_data\\original_distance_spherical_" + str(dimension) +"_500.npy")
    distance_mean[dimension - 2, 0] = np.mean(distance)
    distance = np.load(os.getcwd() + "\\distance_data\\approximate_distance_spherical_" + str(dimension) +"_500.npy")
    distance_mean[dimension - 2, 1] = np.mean(distance)
    distance = np.load(os.getcwd() + "\\generative_distance_data\\generative_distance_spherical_" + str(dimension) +"_500_01.npy")
    distance_mean[dimension - 2, 2] = np.mean(distance)

index = ["2", "3", "4", "5"]
columns = ["Algorithm 1", "Algorithm 2", "Algorithm 3"]
pd.DataFrame(distance_mean, index = index, columns = columns)
```
