```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
plt.subplots(3, 4, figsize=(8, 6))
plt.tight_layout()

index = 1

for sample_style in ["spherical", "degenerate", "discrete"]:

    samples = np.load(os.getcwd() + "\\coefficient_samples\\" + sample_style + "_coefficient_samples_2_2000.npy")
    samples = np.tile(samples, (1, 8))[:, :15900]

    plt.subplot(3, 4, index)
    plt.plot(samples[0, :], samples[1, :], '.', markersize = 0.05)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xticks([])
    plt.yticks([])

    index += 4

index = 2

for sample_style in ["spherical", "degenerate", "discrete"]:

    samples = np.load(os.getcwd() + "\\estimation_data\\original_estimation_" + sample_style + "_2_2000.npy").T.reshape((-1, 2)).T

    plt.subplot(3, 4, index)
    plt.plot(samples[0, :], samples[1, :], '.', markersize = 0.05)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xticks([])
    plt.yticks([])

    index += 4

index = 3

for sample_style in ["spherical", "degenerate", "discrete"]:

    samples = np.load(os.getcwd() + "\\estimation_data\\approximate_estimation_" + sample_style + "_2_2000.npy").T.reshape((-1, 2)).T

    plt.subplot(3, 4, index)
    plt.plot(samples[0, :], samples[1, :], '.', markersize = 0.05)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xticks([])
    plt.yticks([])

    index += 4

index = 4

for sample_style in ["spherical", "degenerate", "discrete"]:

    samples = np.load(os.getcwd() + "\\generative_samples_data\\generative_samples_" + sample_style + "_2_2000_01.npy").T.reshape((-1, 2)).T

    plt.subplot(3, 4, index)
    plt.plot(samples[0, :], samples[1, :], '.', markersize = 0.05)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xticks([])
    plt.yticks([])

    index += 4

plt.subplots_adjust(wspace = 0.05, hspace = 0.05)
plt.savefig("introduction.png")
plt.show()
```

```python
plt.subplots(3, 4, figsize=(8, 6))
plt.tight_layout()

sample_style = ["spherical", "degenerate", "discrete"]

for style_index in range(3):
    for parameter_index in range(4):
        parameter = 2 ** parameter_index
        samples = np.load(os.getcwd() + "\\generative_samples_data\\generative_samples_"+ sample_style[style_index] + "_2_2000_particle_"+ str(parameter) +"0.npy").T.reshape((-1, 2)).T
        samples = np.tile(samples, (1, 8))[:, :8000]

        plt.subplot(3, 4, style_index * 4 + parameter_index + 1)
        plt.plot(samples[0, :], samples[1, :], '.', markersize = 0.25)
        plt.xlim(-10, 10)
        plt.ylim(-10, 10)
        plt.xticks([])
        plt.yticks([])


plt.subplots_adjust(wspace = 0.05, hspace = 0.05)
plt.savefig("particles.png")
plt.show()
```
